```python
import pandas as pd

dataframe = pd.read_csv('exercise3_file.csv')

dataframe["type"] = dataframe.idxmax(axis = 1, skipna = True)

dataframe["avg"]  = (dataframe["math"]*2 + dataframe["eng"]*1.5 + dataframe["chinese"]*1.5)/5

dataframe = dataframe.replace(to_replace = {"math":"A", "eng":"B", "chinese":"C"})

print(dataframe)
```
